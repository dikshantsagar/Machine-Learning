```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
oliv=datasets.fetch_olivetti_faces()
```

```python
fig=plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(oliv.images[i],cmap=plt.cm.bone)
plt.show()
```

```python
pca_oliv=PCA()
x=oliv.data
y=oliv.target
```

```python
pca_oliv.fit(x)
pca_oliv.components_.shape
```

```python
k=0
total=0
while total < 0.99:
    total+=pca_oliv.explained_variance_ratio_[k]
    k+=1
k
```

```python
pca_oliv=PCA(n_components=260)
xt=pca_oliv.fit_transform(x)
```

```python
xt.shape
```

```python
x_app=pca_oliv.inverse_transform(xt)
x_app=x_app.reshape((400,64,64))
```

```python
pca_oliv.components_.shape
```

```python
fig=plt.figure(figsize=(16,16))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(pca_oliv.components_.reshape(260,64,64)[i],cmap=plt.cm.bone)
plt.show()
```
