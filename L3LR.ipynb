```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs=np.array([1,2,3,4,5,6])
ys=np.array([1,2,3,4,5,6])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
xsr=xs.reshape(6,1)
clf=LinearRegression()
clf.fit(xsr,ys)
```

```python
xp=np.array([[0],[3.5],[10]])
yp=clf.predict(xp)
plt.plot(yp)

```
