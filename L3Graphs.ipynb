```python
import matplotlib.pyplot as plt
```

```python
x=[5,8,10]
```

```python
y=[10,16,20]
plt.plot(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,y,Color='r' , label = 'var 1')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Demo")
x1=[15,18,20]
x2=[12,16,20]
plt.plot(x1,x2,Color='r',marker='^', label='var 2')
plt.axis([0,25,0,30])
plt.grid(True)
plt.text(11,11,'HELLO')
plt.legend()
plt.show()
```
