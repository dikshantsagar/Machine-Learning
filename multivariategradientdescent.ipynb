```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import math

#calculate cost funtions
def cost(x,y,m):
    return ((y-np.dot(x,m))**2).mean()

#calculate gradients and updte m and b values
def step_gradient(x,y,learning_rate ,m, batch_size):
    k=x.shape[0]
    n=x.shape[1]
    num_batches=math.ceil(k/batch_size)
    for b in range(num_batches):
        start=b * batch_size
        end= min((b+1)*batch_size , k)
        slope_m=np.zeros((x.shape[1],1))
        for j in range(n):
            
                y_pred_i=np.dot(x,m)
                slope_m+=((2/k)*np.dot((y_pred_i - y).reshape(1,379),x)).reshape(13,1)
                
        m= m -learning_rate*slope_m
    return m
    

#running multiple iterations of step gradient
def gradient_descent(x,y,learning_rate=0.1, num_iter = 100 , batch_size=32):
    m=np.zeros((x.shape[1],1))
    print("Start : ", cost(x,y,m))
    for i in range(num_iter):
        m = step_gradient(x,y,learning_rate ,m, batch_size)
        print(i,": ", cost(x,y,m)) 
    print("END : ", cost(x,y,m))
    return (m)

#data loading and calling gradient discent
def run(x,y):
    y=y.reshape((x.shape[0],1))
    learning_rate=0.01
    num_iter=1000
    m=gradient_descent(x,y, learning_rate, num_iter)
    print(m)
    return m
    

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

clf=LinearRegression(fit_intercept=False)

def score(y,y_pred):
    u=((y-y_pred)**2).sum()
    v=np.var(y)
    return 1-u/v  

btrain=pd.read_csv("~/desktop/codingninjas/btrain.csv")

btest=pd.read_csv("~/desktop/codingninjas/btest.csv")

x=np.asarray(btrain.iloc[:,0:13])

y=np.asarray(btrain.iloc[:,-1:])

clf.fit(x,y)

clf.coef_

m=run(x,y)
```
